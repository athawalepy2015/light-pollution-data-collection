```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
os.makedirs("data", exist_ok = True)
```

```python
r = requests.get("https://globeatnight.org/maps-data/")
```

```python
soup = BeautifulSoup(r.content, "html.parser")
#print(soup)
```

```python
print("Total <a> links found:", len(soup.find_all("a")))
```

```python
urls = []
```

```python
for link in soup.select("li a[href]"):
    href = link["href"]    
    # print(link)
    if (href.endswith(".csv") and "/documents/" in href and "GaN" in href and "20" in href):
        urls.append("http://globeatnight.org" + href)
        
```

```python
with open("data/gan_urls.txt", "w" , encoding = "utf-8") as f:
    for url in urls:
        f.write(url + '\n' )               
```

```python
print("Total CSV files found:", len(urls))
```
